```python
from manim import *
import random
import numpy as np
from sklearn.linear_model import SGDRegressor
```

# Generate Data

```python
data_set_size = 50  # If a larger number is used the model fits data after one epoch.

X = np.random.uniform(-2, 2, size = (data_set_size, 1))

# Adding some noise so that the points won't be exactly on the same line
noise = np.random.normal(-2, 2, size = data_set_size)

y = -3 * X.flatten() + 1 + noise
```

# Train Model

```python
model = SGDRegressor(
    warm_start= True # Reuse the solution of the previous call to fit as initialization, otherwise, just erase the previous solution.
)

epochs = 10
weights_history = []
bias_history = []

for epoch in range(epochs):
    model.partial_fit(X, y) # Train for one epoch
    if epoch%2==0:
      weights_history.append(model.coef_.copy()[0])
      bias_history.append(model.intercept_.copy()[0])

```

# Animate

```python
%%manim -pqm  AnimateLines

class AnimateLines(Scene):

    def construct(self):

        plane = NumberPlane()
        self.add(plane)

        dots = VGroup()
        for i in range(data_set_size):
            a = X[i][0]
            b = y[i]
            dot = Dot(plane.c2p(a, b), radius=0.05, color=BLUE)
            dots.add(dot)
        self.play(FadeIn(dots),run_time = 2)
        self.wait(3)


        a = weights_history[0]
        b = bias_history[0]
        line = plane.plot(lambda x: a*x + b, color=RED)
        self.play(FadeIn(line))
        self.wait(0.5)

        # Animate the transformation to each subsequent line
        for i in range(len(weights_history[1:])):
            a = weights_history[i]
            b = bias_history[i]
            new_line = plane.plot(lambda x: a*x + b, color=RED)
            self.play(Transform(line, new_line))
            self.wait(0.5)
            print(a,b)

```
